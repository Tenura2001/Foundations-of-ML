Helper Functions

```python
def log1p_transform(X):
    return np.log1p(X)

def minmax_transform(X):
    X_min = np.min(X)
    X_max = np.max(X)
    return (X - X_min) / (X_max - X_min)

def standard_transform(X):
    X_mean = np.mean(X)
    X_std = np.std(X)
    return (X - X_mean) / X_std
```
